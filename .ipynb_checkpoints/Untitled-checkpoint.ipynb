```python
rows = int(input("enter number of rows"))
cols = int(input("enter number of columns"))

a = []
b = []
add = []

print("Elements for a")
for i in range(cols):
    row = []
    for j in range(rows):
        n = int(input(f"Enter element for a[{i}][{j}]: "))
        row.append(n)
    a.append(row)

print("Elements for b")
for i in range(cols):
    row = []
    for j in range(rows):
        n = int(input(f"Enter element for b[{i}][{j}]: "))
        row.append(n)
    b.append(row)

for i in range(cols):
    row = []
    for j in range(rows):
        n = a[i][j] + b[i][j]
        row.append(n)
    add.append(row)

for i in add:
    print(i)
```

```python
def fact(n):
    if n < 0:
        return
    if n == 0:
        return 1
    if n == 1:
        return 1
    return n*fact(n-1)

print(fact(6))
```

```python
def my_range(start,end,jump):
    while(start != end):
        yield start
        start += jump

for i in my_range(0,16,1):
    print(i)
        
```

```python
def is_prime(n):
    if n < 2:
        return False
    if n == 2:
        return True
    count = 0
    for i in range(1,n+1):
        if n%i == 0:
            count += 1
    if count == 2:
        return True
    return False

def next_prime(n):
    n += 1
    while not is_prime(n):
        n += 1
    yield n

print(next(next_prime(5)))
print(next(next_prime(6)))
print(next(next_prime(9)))
```

```python
def fib(n):
    if n == 0:
        return 1
    if n == 1:
        return 1
    return fib(n-1)+fib(n-2)

for i in range(5):
    print(fib(i))
        
```

```python
def is_prime(n, i=2):
    if n < 2:
        return False
    if i*i > n:
        return True
    if n%i == 0:
        return False
    return is_prime(n,i+1)

def print_prime(s,e):
    if s > e:
        return
    if is_prime(s):
        print(s)
    return print_prime(s+1,e)

print_prime(5,60)
```

```python
from math import pi

class circle:
    def __init__(self,r):
        self.r = r

    def area(self):
        return pi*(self.r*self.r)

    def perimeter(self):
        return 2*pi*self.r

c = circle(5)
print(c.area())
print(c.perimeter())
```

```python
class bankdemo:
    def __init__(self,acc_no, name, balance):
        self.acc_no = acc_no
        self.name = name
        self.balance = balance

    def deposit(self,amt):
        self.balance += amt
        return

    def withdraw(self,amt):
        self.balance -= amt
        return

    def check_balance(self):
        return self.balance

acc = bankdemo(1,"amit",5000)
acc.deposit(5000)
acc.withdraw(2000)
acc.check_balance()
```

```python
class parent:
    def wclass(self):
        print("this is a parent class")


class child(parent):
    def cclass(self):
        print("hello from child class")
        

c = child()
c.wclass()
c.cclass()
```

```python
class gparent:
    def gph(self):
        print("grandparent")

class parent:
    def pph(self):
        print("parent")

class child(gparent, parent):
    def cph(self):
        print("child")

c = child()
c.gph()
c.pph()
c.cph()
```

```python
class parent:
    def pph(self):
        print("parent")

class child(parent):
    def cph(self):
        print("child1")

class child2(parent):
    def c2ph(self):
        print("child2")
c = child()
c.pph()
c.cph()

c2 = child2()
c2.pph()
c2.c2ph()
```

```python
import re

Url = input("enter a password")
pattern = r'^(https?:\/\/)?(www\.)?[A-Za-z0-9]+\.[a-zA-Z]{2,}(\/\s*)?$'
valid = re.match(pattern, Url)

if(valid):
    print("Valid Url")
else:
    print("Invalid Url")
```

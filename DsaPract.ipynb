```python
rows = int(input("Enter number of rows"))
cols = int(input("Enter number of columns"))

def matrix_add(a, b,rows,cols):
    add = []
    for i in range(rows):
        row = []
        for j in range(cols):
            n = a[i][j] + b[i][j]
            row.append(n)
        add.append(row)
    return add

def matrix_sub(a, b,rows,cols):
    sub = []
    for i in range(rows):
        row = []
        for j in range(cols):
            n = a[i][j] - b[i][j]
            row.append(n)
        sub.append(row)
    return sub

def matrix_mul(a, b,rows,cols):
    mul = []
    for i in range(rows):
        row = []
        for j in range(cols):
            n = a[i][j] * b[i][j]
            row.append(n)
        mul.append(row)
    return mul



A=[]
B=[]
print("input elements of matrix 1")
for i in range(rows):
    row = []
    for j in range(cols):
        n = int(input(f"Enter element at A[{i}][{j}] "))
        row.append(n)
    A.append(row)

print("input elements of matrix 2")
for i in range(rows):
    row = []
    for j in range(cols):
        n = int(input(f"Enter element at B[{i}][{j}] "))
        row.append(n)
    B.append(row)

while True:
    print("MENU :-")
    print(f"1. Add \n2.Subtract \n3.Multiply \n4. Exit" )
    n= int(input("Input option number:- "))
    if n == 1:
        print("Addition:- ",matrix_add(A,B,rows,cols))
    elif n == 2:
        print("Subtraction:- ",matrix_sub(A,B,rows,cols))
    elif n == 3:
        print("Multiplication:- ",matrix_mul(A,B,rows,cols))
    elif n == 4:
        break
    else:
        print("Please Choose appropriate option")
print("Exited Menu
```

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.next=None

class singlyLinkedList:
    def __init__(self):
        self.head = None

    def create(self):
        n = int(input("Enter number of nodes"))
        for i in range(n):
            self.data = int(input(f"Enter element for node {i+1}"))
            self.insert_end(self.data)

    def insert_beginning(self,data):
        new = Node(data)
        new.next = self.head
        self.head = new
        print("Inserted at beginning")

    def insert_end(self,data):
        new = Node(data)
        if self.head is None:
            self.head = new
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new

    def delete(self, value):
        if self.head is None:
            print("List is empty")
            return

        if self.head.data == value:
            self.head = self.head.next
            print(f"Deleted the {value} from linked list")
            return
        temp = self.head
        while temp.next and temp.next.data != value:
            temp = temp.next

        if temp.next is None:
            print("value not found")
        else:
            temp.next = temp.next.next
            print(f"Deleted {value} from the list")
            return

    def search(self, value):
        if self.data is None:
            print("List is empty")
            return
        temp = self.head
        pos = 1
        while temp.next and temp.data != value:
            temp = temp.next
            pos +=1

        if temp.next is None:
            print("Value not found")
            return
        else:
            print(f"Value Found at {pos}")

    def display(self):
        if self.head is None:
            print("List is Empty")
            return

        temp = self.head
        while temp:
            print(temp.data, end=" -----> ")
            temp=temp.next
        print("NULL")


l = singlyLinkedList()
while True:
    print("Menu")
    print("1. Create a Linked List \n2. Insert at beginning \n3. Insert at end \n4. Delete from list \n5. Search \n6. Display \n7. Exit")
    n = int(input("Enter Choice:- "))
    if n == 1:
        l.create()
    elif n == 2:
        a = int(input("Enter Value to be Inserted"))
        l.insert_beginning(a)
    elif n == 3:
        a = int(input("Enter Value to be Inserted at end"))
        l.insert_end(a)
    elif n == 4:
        a = int(input("Enter Value to be deleted"))
        l.delete(a)
    elif n == 5:
        a = int(input("Enter Value to be searched"))
        l.search(a)
    elif n == 6:
        l.display()
    elif n == 7:
        break
    else:
        print("Invalid Choice")
print("Exited the menu")
            
```

```python
class Node:
    def __init__(self,data):
        self.data = data
        self.prev = None
        self.next = None

class DoublyLL:
    def __init__(self):
        self.head = None

    def create(self):
        n = int(input("Enter Number of Nodes:- "))
        for i in range(n):
            data = int(input(f"Enter element for node {i+1}"))
            self.insert_end(data)
        
    def insert_beg(self,data):
        new = Node(data)
        new.next = self.head
        if(self.head):
            self.head.prev = new
        self.head = new

    def insert_end(self,data):
        new = Node(data)
        if self.head is None:
            self.head = new
            return
        temp = self.head
        while temp.next:
            temp = temp.next
        temp.next = new
        new.prev = temp

    def delete(self,value):
        if self.head is None:
            print("list is Empty")
            return

        temp = self.head

        if temp.data == value:
            self.head.next = None
            self.head.prev = None
            self.head = self.head.next
            return

        while temp and temp.data != value:
            temp = temp.next

        if temp.next is None:
            print("Value not Found")
            return
        if temp.next:
            temp.next.prev = temp.prev
        if temp.prev:
            temp.prev.next = temp.next
        print(f"Deleted {value} successfully")

    def search(self,value):
        if self.head is None:
            print("List is empty")
            return

        temp = self.head
        pos = 1

        while temp and temp.data != value:
            temp =  temp.next
            pos += 1

        if temp.next is None:
            print("Value not found")
            return

        print("Found Value at ",pos)

    def display(self):
        if self.head is None:
            print("List is empty")
            return
        temp = self.head
        while temp:
            print(f"[{temp.data}]", end="  <------>  ")
            temp = temp.next
        print("NULL")
        

l = DoublyLL()
while True:
    print("Menu")
    print("1. Create a Linked List \n2. Insert at beginning \n3. Insert at end \n4. Delete from list \n5. Search \n6. Display \n7. Exit")
    n = int(input("Enter Choice:- "))
    if n == 1:
        l.create()
    elif n == 2:
        a = int(input("Enter Value to be Inserted"))
        l.insert_beg(a)
    elif n == 3:
        a = int(input("Enter Value to be Inserted at end"))
        l.insert_end(a)
    elif n == 4:
        a = int(input("Enter Value to be deleted"))
        l.delete(a)
    elif n == 5:
        a = int(input("Enter Value to be searched"))
        l.search(a)
    elif n == 6:
        l.display()
    elif n == 7:
        break
    else:
        print("Invalid Choice")
print("Exited the menu")
```
